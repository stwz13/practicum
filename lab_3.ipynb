```python
def format_matrix(matrix):
    if len(matrix.shape) == 1:
        return " ".join(str(numb) for numb in matrix)
    formatted_matrix = ""
    for row in matrix:
        formatted_row = " ".join(str(numb) for numb in row)
        formatted_matrix += formatted_row + "\n"
    return formatted_matrix
```

```python
import scipy as sc
import numpy as np

matrix = [[4, -3.4, 1, 1.8],
                [-2, 8, 0, -9],
                [2, -2.5, -7, 4],
                [1, -1, 5, 1]]
p, l, u = sc.linalg.lu(matrix)
print(*map(format_matrix, (p,l,u)), sep="\n")

det = np.linalg.det(np.linalg.inv(p)) * np.prod(np.linalg.diagonal(l)) * np.prod(np.linalg.diagonal(u))

print(det)
```

```python
v1 = np.random.uniform(0, 100, 100).astype(int)
v2 = np.random.normal(50, 50, 100).astype(int)
print(*map(format_matrix, (v1, v2)), sep="\n")

```

```python
for v in v1, v2:
    sr = np.mean(v) # среднее
    mod = sc.stats.mode(v)[0] # мода
    median = np.median(v)
    max = np.max(v)
    min = np.min(v)
    std = np.std(v)

    print(f"Среднее: {sr}, мода: {mod}, медиана: {median}"
          f", максимум {max}, минимум: {min}, стандартное отклонение: {std}")



```
