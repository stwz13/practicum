```python
df.drop_duplicates(inplace=True)
df.drop(columns="Unnamed: 12", inplace=True)
df.rename(columns={"game_name":"game"}, inplace=True)

df["highest_price"] = (df["highest_price"]
                       .str.replace("€", "")
                       .str.strip())
df["highest_price"] = pd.to_numeric(df["highest_price"], downcast="float")

df["highest_price"] = df["highest_price"].round()
df["playstation_score"] = df["playstation_score"].replace("--", np.nan)
df["playstation_score"] = pd.to_numeric(df["playstation_score"], downcast="float")

df
```

```python
plt.hist(df["highest_price"], bins=10, color="green")
plt.xlabel("Highest_price")
plt.ylabel("Number of games")
plt.savefig("pictures/histogram.png", bbox_inches = "tight")
plt.show()
```

```python
sb.boxplot(data=df, x=df["highest_price"], color="green")
plt.savefig("pictures/boxplot.png", bbox_inches = "tight")
```

```python
separated_platforms  = df["platform"].dropna().str.split("/").explode().str.strip()

platforms_count = separated_platforms.value_counts()

plt.pie(platforms_count.values, labels = platforms_count.index)
plt.savefig("pictures/pie.png", bbox_inches = "tight")
plt.show()
```

```python
corr_matrix = df.corr(numeric_only = True)
sb.heatmap(corr_matrix, annot=True, cmap="mako")
plt.savefig("pictures/heatmap.png", bbox_inches = "tight")
```

```python
main_genre_df = pd.DataFrame()

main_genre_df["main_genre"]  = df["genre"].dropna().str.split("/").str[0].str.strip()
main_genre_df["main_genre"].replace("--", np.nan)
main_genre_df["newest_platform"] = df["platform"].dropna().str.split("/").str[0].str.strip()

plt.figure(figsize=(10, 10))
sb.countplot(main_genre_df, x = "newest_platform", hue="main_genre")
plt.savefig("pictures/countplot.png", bbox_inches = "tight")
```

```python
missing_data = pd.DataFrame({"missing_data": df.isna().sum()})
missing_data
```

```python
for float_column in ["highest_price", "metacritic_score",
                     "metacritic_rating_count", "metacritic_user_score", "metacritic_user_rating_count",
                     "playstation_rating_count"]:
    df[float_column] = df[float_column].fillna(df[float_column].mean())

for object_column in ["release_date", "genre", "publisher", "platform"]:
    df[object_column] = df[object_column].fillna(df[object_column].mode()[0])

df
```

```python
metacritic_score_sample = df["metacritic_score"][:200]
_, p_value = sc.stats.normaltest(metacritic_score_sample)
print(f"{p_value: .10f}")
```

```python
df["main_genre"] = df['genre'].str.split("/").str[0].str.strip()

platforms = df["platform"].str.split("/").explode().str.strip().unique()
for separated_platform in platforms:
    df[separated_platform] = df["platform"].str.contains(separated_platform)

df
```

```python
preprocessed_df = pd.get_dummies(df, columns=["PS3","PS4", "PS5", "PS Vita", "PSP", "publisher", "main_genre"])
preprocessed_df.to_csv("preprocessed_df.csv")
preprocessed_df
```
