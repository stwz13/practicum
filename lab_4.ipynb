```python
def format_matrix(matrix):
    if len(matrix.shape) == 1:
        return " ".join(str(numb) for numb in matrix)
    formatted_matrix = ""
    for row in matrix:
        formatted_row = " ".join(str(numb) for numb in row)
        formatted_matrix += formatted_row + "\n"
    return formatted_matrix
```

```python
import numpy as np
import matplotlib.pyplot as plt

y = lambda x: np.log(np.sqrt(x))
a = 1
b = 6

n = 100
length = 1.0 * (b - a) / 100
points = np.array([min(a + i * length, b) for i in range(n)])
y_values_in_points = y(points)

plt.plot(points, y_values_in_points, label ="График 1")
plt.xlabel("Ось Ox", fontsize = 15)
plt.ylabel("Ось Oy", fontsize = 15)
plt.title("График функции")
plt.rc('xtick', labelsize = 15)
plt.rc('ytick', labelsize = 15)
plt.legend()

plt.show()
```

```python
plt.scatter(points, y(points), marker ="*", color = (0.1, 0.9, 0.3))
plt.title("График 2")
plt.xlabel("Ось Ox", fontsize = 15)
plt.ylabel("Ось Oy", fontsize = 15)
plt.grid(color = "black", alpha = 0.13)
plt.show()
```

```python
v1 = np.random.uniform(0, 100, 100).astype(int)
v2 = np.random.normal(50, 15, 100).astype(int)

plt.hist(v1, color = "blue", bins = 50)
plt.title("Гистограмма 1")
plt.show()
plt.hist(v2, color = "red", bins = 100)
plt.title("Гистограмма 2")
plt.show()
```

```python
v3 = np.random.uniform(1, 5, 50).astype(int)

unique_values, counts = np.unique(v3, return_counts=True)

plt.bar(unique_values, counts)
plt.title("Столбчатая диаграмма", color="blue")
plt.show()

plt.pie(counts, labels=["1", "2", "3", "4"], colors=["red", "green", "blue", "yellow"])
plt.title("Круговая диаграмма")
plt.show()
```

```python
fig = plt.figure(figsize=(10, 15))
ax_3d = fig.add_subplot(projection='3d')

x_grid, y_grid = np.meshgrid(points, y_values_in_points)
z_grid = (x_grid - 3)**2 + (y_grid - 1)**2
ax_3d.plot_surface(x_grid, y_grid, z_grid, color = "blue")
```

```python
def create_four_plots(x_grid, y_grid, z_grid):
    fig, axs = plt.subplots(2, 2, figsize = [15,10])
    fig.suptitle("Графики")

    axs[0,0].plot(points, y_values_in_points)
    axs[0,0].set_facecolor("grey")
    axs[0,1].scatter(points, y(points), marker ="*", color = (0.1, 0.9, 0.3))
    axs[1,0].pie(counts, labels=["1", "2", "3", "4"], colors=["red", "green", "blue", "yellow"])
    axs[1,1].remove()
    ax_3d = fig.add_subplot(224, projection = '3d')
    ax_3d.plot_surface(x_grid, y_grid, z_grid, color = "blue")

create_four_plots(x_grid, y_grid, z_grid)


```

```python
plt.style.use('ggplot')
create_four_plots(x_grid, y_grid, z_grid)
```
