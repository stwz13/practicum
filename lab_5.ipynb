```python
import pandas as pd
```

```python
my_df = pd.DataFrame({'Name': ["Strom, Mrs. Wilhelm (Elna Matilda Persson)", "Navratil, Mr. Michel (\"Louis MHoffman\")", "Minahan, Miss. Daisy E"],
                   'Age': [29, 36.5, 33],
                   'Sex': ["female", "male", "female"]})
print(my_df)
```

```python
df_from_file = pd.read_csv("titanic_csv.csv", sep = ";")
df_from_url = pd.read_csv("https://gist.githubusercontent.com/zaryanezrya/8b4ef51c707cb16d5e88a44dc00a1bb2/raw/41230f49c6268e072dbf102672f670be256922ab/gistfile1.txt")

df_from_url.columns = df_from_file.columns
df = pd.concat([df_from_url, df_from_file])

df.drop_duplicates(inplace=True)
print(df)
```

```python
df.set_index("PassengerId", inplace = True)
print(df)
df.sort_values(df.index.name, inplace = True)
```

```python
print(df.info)
```

```python
print(df.describe())
```

```python
df.loc[2], df.iloc[0] = df.iloc[0].copy(), df.loc[2].copy()
print(df)
```

```python
replacement_map = {"female": "f", "male": "m"}
df['Sex'] = df['Sex'].map(replacement_map)
print(df)
```

```python
passengers = df[df.groupby("Ticket")["Name"].transform('count') >= 6]

passengers_names = passengers["Name"]

print(passengers_names)
```

```python
cabins_of_selected_passengers = df[df["Name"].isin(my_df["Name"])]["Cabin"]

passengers_in_selected_cabins = df[df["Cabin"].isin(cabins_of_selected_passengers)]
print(passengers_in_selected_cabins)
```
